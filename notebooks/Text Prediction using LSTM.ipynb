# Remove new line characters

```python
from re import sub
Q = sub('\n',' ',Q)
Q = sub((r'\?'),'',Q)
Q = sub((r'  '),' ',Q)

Q
```

# Unify Case

```python
Q = Q.upper()
Q
```

# Tokenzing

```python
Q = Q.split(' ')
Q
```

# convert data in 3-->1 format

```python
inp = []
op  = []

for i in range(3,len(Q),1):
    w1 = Q[i-3]
    w2 = Q[i-2]
    w3 = Q[i-1]
    w4 = Q[i]
    inp.append(w1+' '+w2+' '+w3+' ')
    op.append(w4)
```

```python
inp
```

```python
op
```

# Create dataframe

```python
import pandas as pd
```

```python
Q1 = pd.DataFrame([inp,op]).T
Q1.columns=(['Input','Output'])
Q1
```

# Define X and Y

### Create list of unique words

```python
from numpy import unique
unique_words = list(unique(Q))
unique_words
```

```python
len(unique_words)
```

## Create array X of zeros for input

```python
from numpy import zeros
```

```python
X = zeros((Q1.shape[0],len(unique_words),(Q1.shape[1]+1)))
```

## Fill array with 1's in X

```python
for line_no, line in enumerate(inp):
    for word_no,word in enumerate(line.split()):
        X[line_no,unique_words.index(word),word_no] =1
```

### checking the value

```python
unique_words.index('LAGAANA')
```

```python
X[8,46,2]
```

# Create array Y of zeros for Output

```python
Y = zeros((Q1.shape[0],len(unique_words)))
```

```python
Y
```

## fill array Y with 1's

```python
for line_no, word in enumerate(list(Q1['Output'])):
    Y[line_no,unique_words.index(word)]=1
```

### Checking the value

```python
unique_words.index('RAHA')
```

```python
Y[10,72]
```

# create LSTM Model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
from keras.layers.core import Activation
```

```python
nn = Sequential()
nn.add(LSTM(100,input_shape=(92,3)))
nn.add(Dense(92))
nn.add(Dropout(0.2))
nn.add(Dense(92))
nn.add(Dropout(0.2))
nn.add(Dense(92))
nn.add(Activation('softmax'))
```

```python
nn.compile(loss='categorical_crossentropy',metrics=['accuracy'],optimizer = 'adam')
```

```python
model = nn.fit(X,Y,validation_split=0.1,validation_batch_size=20,epochs = 80)
```

```python
acc =model.history['accuracy']
val_acc= model.history['val_accuracy']
import matplotlib.pyplot as plt
plt.plot(acc)
plt.plot(val_acc)
```
